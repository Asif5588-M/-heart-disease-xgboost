Project: heart-disease-xgboost

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

df = pd.read_csv('../data/raw/YOUR_FILE.csv')
df.head()
```

```python
# Shape and info
print('Shape:', df.shape)
print('\nData Types:\n', df.dtypes)
print('\nMissing Values:\n', df.isnull().sum())
```

```python
# Statistical summary
df.describe()
```

```python
# Target distribution
target_col = 'target'  # CHANGE THIS
df[target_col].value_counts().plot(kind='bar', color=['steelblue','salmon'])
plt.title('Target Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('../data/processed/target_dist.png')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.savefig('../data/processed/correlation.png')
plt.show()
```
